```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "City":["New York","Hyderabad","Toranto"],
    "Temperature":[6,35,2]
})

df1

```

```python
df2 = pd.DataFrame({
    "City":["Hyderabad","New York","Toranto"],
    "Humidity":[32,25,60]
})

df2
```

```python
pd.merge(df1,df2,on='City')
```

```python
df1 = pd.DataFrame({
    "City":["New York","Hyderabad","Pune","Bangalore"],
    "Temperature":[6,35,2,9]
})

df1
```

```python
df2 = pd.DataFrame({
    "City":["Hyderabad","New York","Toranto"],
    "Humidity":[32,25,60]
})

df2
```

```python
pd.merge(df1,df2,on='City')
```

```python
pd.merge(df1,df2,on='City',how='outer')
```

```python
pd.merge(df1,df2,on='City',how='left')
```

```python
pd.merge(df1,df2,on='City',how='right')
```

```python
pd.merge(df1,df2,on='City',how='outer',indicator=True)
```

```python
# suffixes
df1 = pd.DataFrame({
    "City":["New York","Hyderabad","Pune","Bangalore"],
    "Temperature":[6,35,2,9],
     "Humidity":[32,25,60,21]
})
df1
```

```python
df2 = pd.DataFrame({
    "City":["New York","Hyderabad","Pune","Bangalore"],
    "Temperature":[10,45,63,15],
     "Humidity":[45,40,35,23]
})
df2
```

```python
pd.merge(df1,df2,on='City')
```

```python
pd.merge(df1,df2,on='City',suffixes=('_left','_right'))
```
